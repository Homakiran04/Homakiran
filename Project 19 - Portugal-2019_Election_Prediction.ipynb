```python
# Defining 'X' as a independent variable.
X = df.drop('FinalMandates',axis=1)
X
```

```python
# Defining 'y' as a Dependent variable.
y = df['FinalMandates']
y
```

```python
# By using train_test_split we are splitting the data into training and testing model.
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
# By checking with DecisionTreeModel algorithm.
from sklearn.tree import DecisionTreeRegressor
Dt = DecisionTreeRegressor()
Dt.fit(X_train,y_train)
y_pred = Dt.predict(X_test)
```

```python
# Predicted valuess.
y_pred
```

```python
# Score for the training model.
Dt.score(X_train,y_train)
```

```python
# Checking r2_score.
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))
```

```python
# Importing multiple Algorithms and metrics of mean_absolute,mean_squared error and r2_score.
# Importing GridsearchCV and cross validation score.
from sklearn.linear_model import Lasso,Ridge,ElasticNet,LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor

from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.model_selection import GridSearchCV,cross_val_score
```

```python
# Using multiple algorithms to fit and predict the model.
import warnings
warnings.simplefilter('ignore')
model = [SVR(),DecisionTreeRegressor(),KNeighborsRegressor(),AdaBoostRegressor(),GradientBoostingRegressor(),Lasso(),Ridge(),ElasticNet(),LinearRegression()]

for m in model:
    m.fit(X_train,y_train)
    score = m.score(X_train,y_train)
    m_pred = m.predict(X_test)
    print('score of',m,'is:',score)
    print('MAE:',mean_absolute_error(y_test,m_pred))
    print('MSE:', mean_squared_error(y_test,m_pred))
    print('RMSE:', np.sqrt(mean_squared_error(y_test,m_pred)))
    print('R2_score:', r2_score(y_test,m_pred))
    print('******************************************************')
    print('\n')
```

Observation:
    --> All Algorithms performed Model well and Maximum r2_score,fit model is GradientBoostinRegressor.

```python
# Defining Cross_Validation scores.
def get_cv_scores(model):
    scores = cross_val_score(model,
                            X_train,y_train,cv=5,
                            scoring='r2')
    print('CV mean : ', np.mean(scores))
    print('\n')
```

```python
# Cv score for Linear Regression.
get_cv_scores(LinearRegression())
```

```python
# Cv score for DecisionTreeRegressor.
get_cv_scores(DecisionTreeRegressor())
```

```python
# Cv score for GradientBoostingRegressor.
get_cv_scores(GradientBoostingRegressor())
```
