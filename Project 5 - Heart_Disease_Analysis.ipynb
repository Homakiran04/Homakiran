```python
# Dropping the values.
newdf = df.drop(columns = ['trestbps','chol','fbs','restecg','thalach'])
```

```python
# New DataFrame after dropping. 
newdf
```

```python
# Splitting the data into x and y
X = newdf.drop('num',axis=1)
y = newdf['num']
```

```python
# Importing standardscaler,cross-validation,confusion_matrix and classification report
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score,train_test_split
from sklearn.metrics import confusion_matrix,classification_report
```

```python
# Normalizing the X values
normal_X = (X - np.min(X)) / (np.max(X)) - np.min(X)
```

```python
# Training and testing the dataset
X_train,X_test,y_train,y_test = train_test_split(normal_X,y,random_state=0)
print('X_test',X_test.shape)
print('X_training',X_train.shape)
```

```python
# Importing pca to get variance ratio
from sklearn.decomposition import PCA
pca = PCA().fit(X_train)
print(pca.explained_variance_ratio_)
print(X_train.columns.values.tolist())
print(pca.components_)
```

```python
# By Using Decisiontreeclassifier we are fitting the model and predicting the model 
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train,y_train)
dtc_pred = dtc.predict(X_test)
```

```python
# Finding accuracy of the model
dtc_accuracy = dtc.score(X_test,y_test)
```

```python
# Accuracy of the model
dtc_accuracy
```

```python
print('Testing Accuracy of DecisionTreeClassifier Model:',dtc_accuracy)
```
