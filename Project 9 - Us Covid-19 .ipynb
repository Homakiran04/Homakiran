# MultiOutputRegressor Method :

```python
# Using make_regression method with Linear regression algorithm
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression

X,y = make_regression(n_samples=58,n_features=16,n_targets=2,random_state=1,noise=0.5)

Lr = LinearRegression()
Lr.fit(X,y)
```

```python
# Predicting the model
y_pred = Lr.predict(X)
```

```python
# Predicted values.
y_pred
```

```python
# Checking score of the model.
Lr.score(X,y)
```

```python
# Using Multi_output_regressor algorithm to predict the model.
from numpy import absolute
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
from sklearn.multioutput import MultiOutputRegressor
df = MultiOutputRegressor(Lr)
cv = RepeatedKFold(n_splits=10,n_repeats=2,random_state=1)
n_scores = cross_val_score(df,X,y,scoring='neg_mean_absolute_error',cv=cv,n_jobs=1)
n_score = absolute(n_scores)
```

```python
# Absolute scores from cross_validation.
n_score
```

# Linear_Regression model :

```python
# Splitting the data into training and testing model.
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
# Score of training model from multi_output_regressor model.
Lr.fit(X_train,y_train)
Lr.score(X,y)
```

```python
# Predicted values of the Deaths and Recovered.
y_pred = Lr.predict(X_test)
y_pred
```

```python
# Score of the training model.
Lr.score(X_train,y_train)
```

```python
# Finding mean_squared_error and mean_absolute_error,root_mean_squared_error with metrics.
from sklearn.metrics import mean_squared_error,mean_absolute_error
print('Mean absolute error:',mean_absolute_error(y_test,y_pred))
print('Mean squared error :',mean_squared_error(y_test,y_pred))
print('Root Mean squared error:',np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
# R2-score of the model.
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))
```

```python
# Getting cross-validation scores and standard deviation.
scores = cross_val_score(df,X_train,y_train,cv=5,scoring='r2')
scores
print('STD:', np.std(scores))
print('CV Mean:', np.mean(scores))
```
