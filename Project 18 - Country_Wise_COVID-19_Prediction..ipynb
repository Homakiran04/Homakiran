```python
# Label Encoding the values of 'Country' to numerics.
from sklearn.preprocessing import LabelEncoder
Le = LabelEncoder()
df['Country'] = Le.fit_transform(df['Country'])
```

```python
# Checking DataFrame.
df.head()
```

```python
# And, Finally dropping the date column as un_used columns.
df = df.drop('Date',axis=1)
```

```python
# Finding Correlation that how much co-related to each other by columns. 
df.corr()
```

```python
# Importing Zscore to remove Outliers.
from scipy.stats import zscore
z = np.abs(zscore(df))
z
```

```python
# Threshold value should be constant 'three'.
threshold = 3
print(np.where(z>3))
```

```python
# Finally Removing Outliers.
df_new=df[(z<3).all(axis=1)]
```

```python
# After Removing Outliers.
df_new
```

```python
# Again Getting DataFrame into 'df'. 
df = df_new
```

```python
# Checking Head of the data.
df.head()
```

```python
# Defining 'X' variable.
X = df.drop('Total Deaths',axis=1)
X
```

```python
# Defining 'y' variable.
y = df['Total Deaths']
y
```

```python
# Splitting data into training and testing model.
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
# Using one algorithm of DecisionTreeRegressor to check the model.
from sklearn.tree import DecisionTreeRegressor
Dt = DecisionTreeRegressor()
Dt.fit(X_train,y_train)
y_pred = Dt.predict(X_test)
```

```python
# Predicted Values.
y_pred
```

```python
# Decision Tree score for Training model. 
Dt.score(X_train,y_train)
```

```python
# R2 score for the model.
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))
```

```python
# Importing multiple Algorithms and metrics of mean_absolute,mean_squared error and r2_score.
# Importing GridsearchCV and cross validation score.
from sklearn.linear_model import Lasso,Ridge,ElasticNet
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor

from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.model_selection import GridSearchCV,cross_val_score
```

```python
# Using multiple algorithms to fit and predict the model.
model = [SVR(),DecisionTreeRegressor(),KNeighborsRegressor(),AdaBoostRegressor(),GradientBoostingRegressor(),Lasso(),Ridge(),ElasticNet()]

for m in model:
    m.fit(X_train,y_train)
    score = m.score(X_train,y_train)
    m_pred = m.predict(X_test)
    print('score of',m,'is:',score)
    print('MAE:',mean_absolute_error(y_test,m_pred))
    print('MSE:', mean_squared_error(y_test,m_pred))
    print('RMSE:', np.sqrt(mean_squared_error(y_test,m_pred)))
    print('R2_score:', r2_score(y_test,m_pred))
    print('******************************************************')
    print('\n')
```

Observation:
    --> The Best fit model and High Accuracy is KNeighbors Regressor with 99% r2_score.
    
    --> And, the Next best models are DecisionTreeRegressor,GradientBoostingRegressor.
    
    --> And, Lasso,Ridge and ElasticNet models are also did well.

```python
# Defining Cross_Validation scores.
def get_cv_scores(model):
    scores = cross_val_score(model,
                            X_train,y_train,cv=5,
                            scoring='r2')
    print('CV mean : ', np.mean(scores))
    print('\n')
```

```python
# Cv score for DecisionTreeRegressor.
get_cv_scores(DecisionTreeRegressor())
```

```python
# Cv score for KNeighborsRegressor.
get_cv_scores(KNeighborsRegressor())
```
