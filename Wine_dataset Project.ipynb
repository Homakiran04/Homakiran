```python
# x and y values of the data
X = df.drop(['Ash'],axis=1)
Y = df.iloc[:,:1]
```

```python
# training and testing the datasets
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)

print(X_train.shape)
print(X_test.shape)
```

```python
# machine learning algorithms to predict the model
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
models = []

models.append(("Logistic Regression:",LogisticRegression()))
models.append(("Decision Tree:",DecisionTreeClassifier()))
models.append(("Support Vector Machine-linear:",SVC(kernel="linear")))
print('models appended...')
```

```python
# finding Accuracy of the model
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

results = []
names = []
for name,model in models:
    kfold = KFold(n_splits=10, random_state=0)
    cv_result = cross_val_score(model,X_train,Y_train.values.ravel(), cv = kfold,scoring = "accuracy")
    names.append(name)
    results.append(cv_result)
for i in range(len(names)):
    print(names[i],results[i].mean()*100)
```
